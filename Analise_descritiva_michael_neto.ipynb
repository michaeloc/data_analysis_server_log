```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import ast
from tqdm import tqdm
sns.set()

from pyathena import connect
```

```python
conn = connect(s3_staging_dir='s3://aws-athena-query-results-415718651610-us-east-1', region_name='us-east-1')
df = pd.read_sql("SELECT type,CAST(item[1] as JSON) as item,CAST(host[1] as JSON) as host, CAST(trigger[1] as JSON) as trigger, CAST(event[1] as JSON) as event, usc_das_id  FROM das.event limit 30000", conn)
```

```python
df.dropna(inplace=True)
```

```python
dict_table = {x:[] for x in ['type','item_value','item_id','item_desc','items_name','host_add',
                        'host_name','host_com','tigger_severity','t_descr','t_nseverity',
                        't_hostgroup','t_value','t_id','t_name',
                        'e_status','e_severity','e_tags','e_time',
                        'e_date','e_id','e_name']}

# dict_table = {x:[] for x in ['type','item_id','items_name','t_id','e_status','e_severity','e_time','e_date','e_name','e_id']}
for idx in tqdm(df.index):
    
    dict_table['type'].append(df.at[idx,'type'])
    
    dict_table['item_value'].append(df.at[idx,'item'][0])
    dict_table['item_id'].append(df.at[idx,'item'][1])
    dict_table['item_desc'].append(df.at[idx,'item'][2])
    dict_table['items_name'].append(df.at[idx,'item'][3])
    
    dict_table['host_add'].append(df.at[idx,'host'][0])
    dict_table['host_name'].append(df.at[idx,'host'][1])
    dict_table['host_com'].append(df.at[idx,'host'][2])
    
    dict_table['tigger_severity'].append(df.at[idx,'trigger'][0])
    dict_table['t_descr'].append(df.at[idx,'trigger'][1])
    dict_table['t_nseverity'].append(df.at[idx,'trigger'][2])
    dict_table['t_hostgroup'].append(df.at[idx,'trigger'][3])
    dict_table['t_value'].append(df.at[idx,'trigger'][4])
    dict_table['t_id'].append(df.at[idx,'trigger'][5])
    dict_table['t_name'].append(df.at[idx,'trigger'][6])
    
    dict_table['e_status'].append(df.at[idx,'event'][0])
    dict_table['e_severity'].append(df.at[idx,'event'][1])
    dict_table['e_tags'].append(df.at[idx,'event'][2])
    dict_table['e_time'].append(df.at[idx,'event'][3])
    dict_table['e_date'].append(df.at[idx,'event'][4])
    dict_table['e_id'].append(df.at[idx,'event'][5])
    dict_table['e_name'].append(df.at[idx,'event'][6])
```

```python
table = pd.DataFrame(dict_table)
```
